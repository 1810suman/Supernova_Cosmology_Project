```python
# üì¶ Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.integrate import quad
from astropy.constants import c
from astropy import units as u
```

```python
# üìÇ Load dataset
file_path = r"C:/Users/suman/Downloads/Pantheon+SH0ES.txt"
df = pd.read_csv(file_path, sep=r"\s+", comment="#")

# üßπ Clean data
df = df[['zHD', 'MU_SH0ES', 'MU_SH0ES_ERR_DIAG']].dropna()
z = df['zHD'].values
mu_obs = df['MU_SH0ES'].values
mu_err = df['MU_SH0ES_ERR_DIAG'].values
```

```python
# üßÆ Cosmological Model
def E(z, Omega_m):
    return np.sqrt(Omega_m * (1 + z)**3 + (1 - Omega_m))

def luminosity_distance(z, H0, Omega_m):
    """Returns luminosity distance in Mpc"""
    try:
        integral, _ = quad(lambda z_: 1.0 / E(z_, Omega_m), 0, z, limit=200)
        H0_si = (H0 * u.km / u.s / u.Mpc)
        d_L = (c / H0_si).to(u.Mpc).value * (1 + z) * integral
        return d_L
    except Exception:
        return np.nan

def mu_theory(z, H0, Omega_m):
    z = np.atleast_1d(z)
    d_L_array = [luminosity_distance(zi, H0, Omega_m) for zi in z]
    return 5 * np.log10(d_L_array) + 25
```

```python
# üîß Fit H0 and Omega_m
p0 = [70, 0.3]
bounds = ([50, 0.1], [90, 0.5])
popt, pcov = curve_fit(mu_theory, z, mu_obs, sigma=mu_err, p0=p0, bounds=bounds)
H0_fit, Omega_m_fit = popt
H0_err, Omega_m_err = np.sqrt(np.diag(pcov))

print(f"‚úÖ Fitted H‚ÇÄ = {H0_fit:.2f} ¬± {H0_err:.2f} km/s/Mpc")
print(f"‚úÖ Fitted Œ©‚Çò = {Omega_m_fit:.3f} ¬± {Omega_m_err:.3f}")

# ‚è≥ Age of Universe
def age_of_universe(H0, Omega_m):
    integral, _ = quad(lambda z_: 1.0 / ((1 + z_) * E(z_, Omega_m)), 0, np.inf, limit=200)
    H0_si = (H0 * u.km / u.s / u.Mpc).to(1 / u.s)
    t0 = integral / H0_si.value
    return t0 / (60 * 60 * 24 * 365.25 * 1e9)  # Convert to Gyr

t0 = age_of_universe(H0_fit, Omega_m_fit)
print(f"üï∞Ô∏è Estimated Age of the Universe: {t0:.2f} Gyr")
```

```python
# üìà Hubble Diagram
plt.figure(figsize=(9, 6))
plt.scatter(z, mu_obs, s=10, label='Observed', alpha=0.7)
plt.plot(z, mu_theory(z, H0_fit, Omega_m_fit), color='red', label=f"Fit H‚ÇÄ={H0_fit:.1f}, Œ©‚Çò={Omega_m_fit:.2f}")
plt.xscale('log')
plt.xlabel('Redshift (z)')
plt.ylabel('Distance Modulus (Œº)')
plt.title('Hubble Diagram')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# üìä Residuals
mu_model = mu_theory(z, H0_fit, Omega_m_fit)
residuals = mu_obs - mu_model

plt.figure(figsize=(9, 4))
plt.scatter(z, residuals, s=10, alpha=0.7)
plt.axhline(0, color='gray', linestyle='--')
plt.xscale('log')
plt.xlabel('Redshift (z)')
plt.ylabel('Residual (Œº_observed - Œº_model)')
plt.title('Hubble Residuals')
plt.grid(True)
plt.show()
```

```python
# üîß Fit with Fixed Œ©‚Çò = 0.3
def mu_fixed_Om(z, H0):
    return mu_theory(z, H0, Omega_m=0.3)

H0_fixed, cov_fixed = curve_fit(mu_fixed_Om, z, mu_obs, sigma=mu_err, p0=[70])
H0_fixed_val = H0_fixed[0]
H0_fixed_err = np.sqrt(np.diag(cov_fixed))[0]

print(f"üîß Fit with fixed Œ©‚Çò=0.3 ‚Üí H‚ÇÄ = {H0_fixed_val:.2f} ¬± {H0_fixed_err:.2f} km/s/Mpc")
```

```python
# üîç Low-z vs High-z Split
z_split = 0.1
low_mask = z < z_split
high_mask = z >= z_split

# Low-z Fit
H0_low, cov_low = curve_fit(mu_fixed_Om, z[low_mask], mu_obs[low_mask], sigma=mu_err[low_mask], p0=[70])
H0_low_val = H0_low[0]
H0_low_err = np.sqrt(np.diag(cov_low))[0]

# High-z Fit
H0_high, cov_high = curve_fit(mu_fixed_Om, z[high_mask], mu_obs[high_mask], sigma=mu_err[high_mask], p0=[70])
H0_high_val = H0_high[0]
H0_high_err = np.sqrt(np.diag(cov_high))[0]

print(f"üîç Low-z (z < {z_split}): H‚ÇÄ = {H0_low_val:.2f} ¬± {H0_low_err:.2f} km/s/Mpc")
print(f"üîç High-z (z ‚â• {z_split}): H‚ÇÄ = {H0_high_val:.2f} ¬± {H0_high_err:.2f} km/s/Mpc")
```
